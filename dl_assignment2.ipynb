```python
"""
Legal Clause Similarity Detection
Deep Learning Assignment 2 - NUCES FAST

This module implements two baseline architectures for semantic similarity detection:
1. BiLSTM-based Siamese Network
2. Attention-based Encoder Network

Author: Assignment Implementation
Date: November 2025
"""

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from collections import Counter
import re
import time
from typing import Tuple, List, Dict
import warnings
warnings.filterwarnings('ignore')

# TensorFlow/Keras imports
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, optimizers, callbacks
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, roc_curve, precision_recall_curve, auc,
    confusion_matrix, classification_report
)

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# GPU Configuration
def setup_gpu():
    """Configure GPU settings and display available devices"""
    print("="*80)
    print("GPU CONFIGURATION")
    print("="*80)

    # List all available GPUs
    gpus = tf.config.list_physical_devices('GPU')

    if gpus:
        print(f"✓ Found {len(gpus)} GPU(s):")
        for i, gpu in enumerate(gpus):
            print(f"  GPU {i}: {gpu.name}")
            # Enable memory growth to avoid allocating all memory at once
            try:
                tf.config.experimental.set_memory_growth(gpu, True)
                print(f"    Memory growth enabled")
            except RuntimeError as e:
                print(f"    Error setting memory growth: {e}")

        # Set GPU as preferred device
        print("\n✓ GPU will be used for training")
        print(f"  TensorFlow version: {tf.__version__}")

        # Display GPU details
        try:
            for gpu in gpus:
                details = tf.config.experimental.get_device_details(gpu)
                if details:
                    print(f"  GPU Details: {details}")
        except:
            pass
    else:
        print("⚠ No GPU detected. Training will use CPU.")
        print("  Note: Training may be slower on CPU.")

    print("="*80)
    print()

# Setup GPU
setup_gpu()

# Configuration
class Config:
    """Configuration class for model parameters"""
    # Data paths
    DATA_DIR = "archive"

    # Text preprocessing
    MAX_VOCAB_SIZE = 10000
    MAX_SEQUENCE_LENGTH = 200
    EMBEDDING_DIM = 128

    # Model architecture
    LSTM_UNITS = 128
    DENSE_UNITS = 64
    DROPOUT_RATE = 0.3

    # Training
    BATCH_SIZE = 32
    EPOCHS = 20
    VALIDATION_SPLIT = 0.2
    TEST_SPLIT = 0.1

    # Pair generation
    POSITIVE_PAIRS_PER_CATEGORY = 50  # Pairs from same category
    NEGATIVE_PAIRS_PER_CATEGORY = 50  # Pairs from different categories


class DataLoader:
    """Handles loading and preprocessing of legal clause data"""

    def __init__(self, data_dir: str = Config.DATA_DIR):
        self.data_dir = Path(data_dir)
        self.clauses = []
        self.labels = []
        self.category_to_clauses = {}

    def load_data(self) -> Tuple[List[str], List[str]]:
        """Load all CSV files and extract clauses"""
        print("Loading dataset...")
        csv_files = list(self.data_dir.glob("*.csv"))
        print(f"Found {len(csv_files)} CSV files")

        for csv_file in csv_files:
            try:
                df = pd.read_csv(csv_file)
                if 'clause_text' in df.columns and 'clause_type' in df.columns:
                    category = df['clause_type'].iloc[0]

                    # Store clauses by category
                    if category not in self.category_to_clauses:
                        self.category_to_clauses[category] = []

                    for _, row in df.iterrows():
                        clause_text = str(row['clause_text']).strip()
                        if len(clause_text) > 10:  # Filter very short clauses
                            self.clauses.append(clause_text)
                            self.labels.append(category)
                            self.category_to_clauses[category].append(clause_text)
            except Exception as e:
                print(f"Error loading {csv_file}: {e}")
                continue

        print(f"Loaded {len(self.clauses)} clauses from {len(self.category_to_clauses)} categories")
        return self.clauses, self.labels

    def clean_text(self, text: str) -> str:
        """Clean and normalize text"""
        # Remove extra whitespace
        text = re.sub(r'\s+', ' ', text)
        # Remove special characters but keep punctuation
        text = re.sub(r'[^\w\s.,;:()\[\]{}"\'-]', '', text)
        return text.strip()

    def create_pairs(self, positive_pairs_per_cat: int = Config.POSITIVE_PAIRS_PER_CATEGORY,
                    negative_pairs_per_cat: int = Config.NEGATIVE_PAIRS_PER_CATEGORY) -> Tuple[List, List]:
        """Create positive (similar) and negative (dissimilar) pairs"""
        print("Creating clause pairs...")

        positive_pairs = []
        negative_pairs = []

        categories = list(self.category_to_clauses.keys())

        # Create positive pairs (same category)
        for category in categories:
            clauses = self.category_to_clauses[category]
            if len(clauses) < 2:
                continue

            # Create pairs within same category
            np.random.shuffle(clauses)
            pairs_created = 0
            for i in range(len(clauses)):
                if pairs_created >= positive_pairs_per_cat:
                    break
                for j in range(i + 1, len(clauses)):
                    if pairs_created >= positive_pairs_per_cat:
                        break
                    positive_pairs.append((clauses[i], clauses[j], 1))
                    pairs_created += 1

        # Create negative pairs (different categories)
        for category in categories:
            clauses = self.category_to_clauses[category]
            if len(clauses) == 0:
                continue

            pairs_created = 0
            other_categories = [c for c in categories if c != category]

            for _ in range(negative_pairs_per_cat):
                if pairs_created >= negative_pairs_per_cat:
                    break

                # Random clause from current category
                clause1 = np.random.choice(clauses)

                # Random clause from different category
                other_category = np.random.choice(other_categories)
                other_clauses = self.category_to_clauses[other_category]
                if len(other_clauses) > 0:
                    clause2 = np.random.choice(other_clauses)
                    negative_pairs.append((clause1, clause2, 0))
                    pairs_created += 1

        print(f"Created {len(positive_pairs)} positive pairs and {len(negative_pairs)} negative pairs")
        return positive_pairs, negative_pairs


class TextPreprocessor:
    """Handles text tokenization and sequence preparation"""

    def __init__(self, max_vocab_size: int = Config.MAX_VOCAB_SIZE,
                 max_seq_length: int = Config.MAX_SEQUENCE_LENGTH):
        self.max_vocab_size = max_vocab_size
        self.max_seq_length = max_seq_length
        self.tokenizer = None
        self.vocab_size = 0

    def fit_tokenizer(self, texts: List[str]):
        """Fit tokenizer on training texts"""
        print("Fitting tokenizer...")
        self.tokenizer = Tokenizer(num_words=self.max_vocab_size, oov_token="<OOV>")
        self.tokenizer.fit_on_texts(texts)
        self.vocab_size = min(len(self.tokenizer.word_index) + 1, self.max_vocab_size)
        print(f"Vocabulary size: {self.vocab_size}")

    def texts_to_sequences(self, texts: List[str]) -> np.ndarray:
        """Convert texts to padded sequences"""
        sequences = self.tokenizer.texts_to_sequences(texts)
        padded = pad_sequences(sequences, maxlen=self.max_seq_length, padding='post', truncating='post')
        return padded


class BiLSTMSiameseNetwork:
    """BiLSTM-based Siamese Network for semantic similarity"""

    def __init__(self, vocab_size: int, embedding_dim: int = Config.EMBEDDING_DIM,
                 lstm_units: int = Config.LSTM_UNITS, max_seq_length: int = Config.MAX_SEQUENCE_LENGTH):
        self.vocab_size = vocab_size
        self.embedding_dim = embedding_dim
        self.lstm_units = lstm_units
        self.max_seq_length = max_seq_length
        self.model = None

    def build_model(self):
        """Build the Siamese BiLSTM model"""
        # Input layers
        input1 = layers.Input(shape=(self.max_seq_length,), name='input1')
        input2 = layers.Input(shape=(self.max_seq_length,), name='input2')

        # Shared embedding layer
        embedding_layer = layers.Embedding(
            self.vocab_size,
            self.embedding_dim,
            input_length=self.max_seq_length,
            name='embedding'
        )

        # Shared BiLSTM encoder
        lstm_layer = layers.Bidirectional(
            layers.LSTM(self.lstm_units, return_sequences=False),
            name='bilstm'
        )

        # Encode both inputs
        embedded1 = embedding_layer(input1)
        embedded2 = embedding_layer(input2)

        encoded1 = lstm_layer(embedded1)
        encoded2 = lstm_layer(embedded2)

        # Add dropout
        encoded1 = layers.Dropout(Config.DROPOUT_RATE)(encoded1)
        encoded2 = layers.Dropout(Config.DROPOUT_RATE)(encoded2)

        # Compute similarity features
        # Element-wise difference and product
        diff = layers.Subtract()([encoded1, encoded2])
        product = layers.Multiply()([encoded1, encoded2])

        # Concatenate features
        merged = layers.Concatenate()([encoded1, encoded2, diff, product])

        # Dense layers
        dense1 = layers.Dense(Config.DENSE_UNITS, activation='relu', name='dense1')(merged)
        dense1 = layers.Dropout(Config.DROPOUT_RATE)(dense1)
        dense2 = layers.Dense(Config.DENSE_UNITS // 2, activation='relu', name='dense2')(dense1)
        dense2 = layers.Dropout(Config.DROPOUT_RATE)(dense2)

        # Output layer
        output = layers.Dense(1, activation='sigmoid', name='output')(dense2)

        # Create model
        self.model = models.Model(inputs=[input1, input2], outputs=output)

        return self.model

    def compile_model(self, learning_rate: float = 0.001):
        """Compile the model"""
        self.model.compile(
            optimizer=optimizers.Adam(learning_rate=learning_rate),
            loss='binary_crossentropy',
            metrics=['accuracy']
        )

    def get_model_summary(self) -> str:
        """Get model architecture summary"""
        from io import StringIO
        import sys
        old_stdout = sys.stdout
        sys.stdout = buffer = StringIO()
        self.model.summary()
        sys.stdout = old_stdout
        return buffer.getvalue()


class AttentionEncoderNetwork:
    """Attention-based Encoder Network for semantic similarity"""

    def __init__(self, vocab_size: int, embedding_dim: int = Config.EMBEDDING_DIM,
                 lstm_units: int = Config.LSTM_UNITS, max_seq_length: int = Config.MAX_SEQUENCE_LENGTH):
        self.vocab_size = vocab_size
        self.embedding_dim = embedding_dim
        self.lstm_units = lstm_units
        self.max_seq_length = max_seq_length
        self.model = None

    def attention_layer(self, inputs, key_dim=None, name_suffix=''):
        """Self-attention mechanism"""
        if key_dim is None:
            # BiLSTM outputs 2 * lstm_units features
            key_dim = max((self.lstm_units * 2) // 4, 32)

        attention = layers.MultiHeadAttention(
            num_heads=4,
            key_dim=key_dim,
            name=f'attention_{name_suffix}' if name_suffix else 'attention'
        )
        attended = attention(inputs, inputs)
        return attended

    def build_model(self):
        """Build the Attention-based Encoder model"""
        # Input layers
        input1 = layers.Input(shape=(self.max_seq_length,), name='input1')
        input2 = layers.Input(shape=(self.max_seq_length,), name='input2')

        # Shared embedding layer
        embedding_layer = layers.Embedding(
            self.vocab_size,
            self.embedding_dim,
            input_length=self.max_seq_length,
            name='embedding'
        )

        # Shared BiLSTM encoder
        lstm_layer = layers.Bidirectional(
            layers.LSTM(self.lstm_units, return_sequences=True),
            name='bilstm'
        )

        # Encode both inputs
        embedded1 = embedding_layer(input1)
        embedded2 = embedding_layer(input2)

        lstm_out1 = lstm_layer(embedded1)
        lstm_out2 = lstm_layer(embedded2)

        # Apply attention (BiLSTM outputs 2 * lstm_units features)
        key_dim = max((self.lstm_units * 2) // 4, 32)
        attended1 = self.attention_layer(lstm_out1, key_dim=key_dim, name_suffix='1')
        attended2 = self.attention_layer(lstm_out2, key_dim=key_dim, name_suffix='2')

        # Global average pooling
        pooled1 = layers.GlobalAveragePooling1D()(attended1)
        pooled2 = layers.GlobalAveragePooling1D()(attended2)

        # Add dropout
        pooled1 = layers.Dropout(Config.DROPOUT_RATE)(pooled1)
        pooled2 = layers.Dropout(Config.DROPOUT_RATE)(pooled2)

        # Compute similarity features
        diff = layers.Subtract()([pooled1, pooled2])
        product = layers.Multiply()([pooled1, pooled2])

        # Concatenate features
        merged = layers.Concatenate()([pooled1, pooled2, diff, product])

        # Dense layers
        dense1 = layers.Dense(Config.DENSE_UNITS, activation='relu', name='dense1')(merged)
        dense1 = layers.Dropout(Config.DROPOUT_RATE)(dense1)
        dense2 = layers.Dense(Config.DENSE_UNITS // 2, activation='relu', name='dense2')(dense1)
        dense2 = layers.Dropout(Config.DROPOUT_RATE)(dense2)

        # Output layer
        output = layers.Dense(1, activation='sigmoid', name='output')(dense2)

        # Create model
        self.model = models.Model(inputs=[input1, input2], outputs=output)

        return self.model

    def compile_model(self, learning_rate: float = 0.001):
        """Compile the model"""
        self.model.compile(
            optimizer=optimizers.Adam(learning_rate=learning_rate),
            loss='binary_crossentropy',
            metrics=['accuracy']
        )

    def get_model_summary(self) -> str:
        """Get model architecture summary"""
        from io import StringIO
        import sys
        old_stdout = sys.stdout
        sys.stdout = buffer = StringIO()
        self.model.summary()
        sys.stdout = old_stdout
        return buffer.getvalue()


class ModelTrainer:
    """Handles model training and evaluation"""

    def __init__(self, model, model_name: str):
        self.model = model
        self.model_name = model_name
        self.history = None

    def train(self, X_train: Tuple, y_train: np.ndarray,
              X_val: Tuple, y_val: np.ndarray,
              epochs: int = Config.EPOCHS,
              batch_size: int = Config.BATCH_SIZE) -> Dict:
        """Train the model"""
        print(f"\nTraining {self.model_name}...")

        # Callbacks
        early_stopping = callbacks.EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True,
            verbose=1
        )

        reduce_lr = callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=3,
            min_lr=1e-6,
            verbose=1
        )

        # Train
        start_time = time.time()
        self.history = self.model.fit(
            X_train,
            y_train,
            validation_data=(X_val, y_val),
            epochs=epochs,
            batch_size=batch_size,
            callbacks=[early_stopping, reduce_lr],
            verbose=1
        )
        training_time = time.time() - start_time

        print(f"Training completed in {training_time:.2f} seconds")

        return {
            'history': self.history.history,
            'training_time': training_time
        }

    def evaluate(self, X_test: Tuple, y_test: np.ndarray) -> Dict:
        """Evaluate the model"""
        print(f"\nEvaluating {self.model_name}...")

        # Predictions
        y_pred_proba = self.model.predict(X_test, verbose=0)
        y_pred = (y_pred_proba > 0.5).astype(int).flatten()

        # Metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, zero_division=0)
        recall = recall_score(y_test, y_pred, zero_division=0)
        f1 = f1_score(y_test, y_pred, zero_division=0)

        # ROC-AUC
        try:
            roc_auc = roc_auc_score(y_test, y_pred_proba)
        except:
            roc_auc = 0.0

        # PR-AUC
        try:
            precision_curve, recall_curve, _ = precision_recall_curve(y_test, y_pred_proba)
            pr_auc = auc(recall_curve, precision_curve)
        except:
            pr_auc = 0.0

        # Confusion matrix
        cm = confusion_matrix(y_test, y_pred)

        results = {
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'roc_auc': roc_auc,
            'pr_auc': pr_auc,
            'confusion_matrix': cm,
            'y_pred': y_pred,
            'y_pred_proba': y_pred_proba.flatten(),
            'y_true': y_test
        }

        print(f"Accuracy: {accuracy:.4f}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall: {recall:.4f}")
        print(f"F1-Score: {f1:.4f}")
        print(f"ROC-AUC: {roc_auc:.4f}")
        print(f"PR-AUC: {pr_auc:.4f}")

        return results

    def plot_training_history(self, save_path: str = None):
        """Plot training history"""
        if self.history is None:
            return

        fig, axes = plt.subplots(1, 2, figsize=(15, 5))

        # Loss
        axes[0].plot(self.history.history['loss'], label='Training Loss')
        axes[0].plot(self.history.history['val_loss'], label='Validation Loss')
        axes[0].set_title(f'{self.model_name} - Loss')
        axes[0].set_xlabel('Epoch')
        axes[0].set_ylabel('Loss')
        axes[0].legend()
        axes[0].grid(True)

        # Accuracy
        axes[1].plot(self.history.history['accuracy'], label='Training Accuracy')
        axes[1].plot(self.history.history['val_accuracy'], label='Validation Accuracy')
        axes[1].set_title(f'{self.model_name} - Accuracy')
        axes[1].set_xlabel('Epoch')
        axes[1].set_ylabel('Accuracy')
        axes[1].legend()
        axes[1].grid(True)

        plt.tight_layout()

        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            print(f"Training history saved to {save_path}")

        plt.show()


def create_qualitative_examples(model, preprocessor, test_pairs: List,
                                num_examples: int = 5) -> Dict:
    """Create qualitative examples of correct and incorrect predictions"""
    print("\nGenerating qualitative examples...")

    # Prepare test data
    test_texts1 = [pair[0] for pair in test_pairs]
    test_texts2 = [pair[1] for pair in test_pairs]
    test_labels = [pair[2] for pair in test_pairs]

    seq1 = preprocessor.texts_to_sequences(test_texts1)
    seq2 = preprocessor.texts_to_sequences(test_texts2)

    # Predictions
    predictions = model.predict([seq1, seq2], verbose=0)
    pred_labels = (predictions > 0.5).astype(int).flatten()

    # Find correct and incorrect predictions
    correct_indices = []
    incorrect_indices = []

    for i in range(len(test_labels)):
        if pred_labels[i] == test_labels[i]:
            correct_indices.append(i)
        else:
            incorrect_indices.append(i)

    # Sample examples
    correct_examples = []
    incorrect_examples = []

    np.random.shuffle(correct_indices)
    np.random.shuffle(incorrect_indices)

    for idx in correct_indices[:num_examples]:
        correct_examples.append({
            'clause1': test_texts1[idx][:200] + '...' if len(test_texts1[idx]) > 200 else test_texts1[idx],
            'clause2': test_texts2[idx][:200] + '...' if len(test_texts2[idx]) > 200 else test_texts2[idx],
            'true_label': test_labels[idx],
            'predicted_label': pred_labels[idx],
            'confidence': float(predictions[idx][0])
        })

    for idx in incorrect_indices[:num_examples]:
        incorrect_examples.append({
            'clause1': test_texts1[idx][:200] + '...' if len(test_texts1[idx]) > 200 else test_texts1[idx],
            'clause2': test_texts2[idx][:200] + '...' if len(test_texts2[idx]) > 200 else test_texts2[idx],
            'true_label': test_labels[idx],
            'predicted_label': pred_labels[idx],
            'confidence': float(predictions[idx][0])
        })

    return {
        'correct': correct_examples,
        'incorrect': incorrect_examples
    }


def print_qualitative_examples(examples: Dict, model_name: str):
    """Print qualitative examples"""
    print(f"\n{'='*80}")
    print(f"Qualitative Examples - {model_name}")
    print(f"{'='*80}")

    print("\n✓ CORRECTLY PREDICTED PAIRS:")
    print("-" * 80)
    for i, ex in enumerate(examples['correct'], 1):
        print(f"\nExample {i}:")
        print(f"  Label: {'Similar' if ex['true_label'] == 1 else 'Dissimilar'}")
        print(f"  Confidence: {ex['confidence']:.4f}")
        print(f"  Clause 1: {ex['clause1']}")
        print(f"  Clause 2: {ex['clause2']}")

    print("\n✗ INCORRECTLY PREDICTED PAIRS:")
    print("-" * 80)
    for i, ex in enumerate(examples['incorrect'], 1):
        print(f"\nExample {i}:")
        print(f"  True Label: {'Similar' if ex['true_label'] == 1 else 'Dissimilar'}")
        print(f"  Predicted: {'Similar' if ex['predicted_label'] == 1 else 'Dissimilar'}")
        print(f"  Confidence: {ex['confidence']:.4f}")
        print(f"  Clause 1: {ex['clause1']}")
        print(f"  Clause 2: {ex['clause2']}")


def compare_models(results1: Dict, results2: Dict, model1_name: str, model2_name: str):
    """Compare two models and print comparison"""
    print(f"\n{'='*80}")
    print("MODEL COMPARISON")
    print(f"{'='*80}")

    metrics = ['accuracy', 'precision', 'recall', 'f1_score', 'roc_auc', 'pr_auc']

    comparison_data = {
        'Metric': metrics,
        model1_name: [results1[m] for m in metrics],
        model2_name: [results2[m] for m in metrics]
    }

    df = pd.DataFrame(comparison_data)
    print("\nPerformance Metrics:")
    print(df.to_string(index=False))

    # Training time comparison
    print(f"\nTraining Time:")
    print(f"  {model1_name}: {results1.get('training_time', 0):.2f} seconds")
    print(f"  {model2_name}: {results2.get('training_time', 0):.2f} seconds")


def plot_roc_curves(results1: Dict, results2: Dict, model1_name: str, model2_name: str):
    """Plot ROC curves for both models"""
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # ROC Curve
    fpr1, tpr1, _ = roc_curve(results1['y_true'], results1['y_pred_proba'])
    fpr2, tpr2, _ = roc_curve(results2['y_true'], results2['y_pred_proba'])

    axes[0].plot(fpr1, tpr1, label=f'{model1_name} (AUC={results1["roc_auc"]:.4f})')
    axes[0].plot(fpr2, tpr2, label=f'{model2_name} (AUC={results2["roc_auc"]:.4f})')
    axes[0].plot([0, 1], [0, 1], 'k--', label='Random')
    axes[0].set_xlabel('False Positive Rate')
    axes[0].set_ylabel('True Positive Rate')
    axes[0].set_title('ROC Curves')
    axes[0].legend()
    axes[0].grid(True)

    # Precision-Recall Curve
    precision1, recall1, _ = precision_recall_curve(results1['y_true'], results1['y_pred_proba'])
    precision2, recall2, _ = precision_recall_curve(results2['y_true'], results2['y_pred_proba'])

    axes[1].plot(recall1, precision1, label=f'{model1_name} (AUC={results1["pr_auc"]:.4f})')
    axes[1].plot(recall2, precision2, label=f'{model2_name} (AUC={results2["pr_auc"]:.4f})')
    axes[1].set_xlabel('Recall')
    axes[1].set_ylabel('Precision')
    axes[1].set_title('Precision-Recall Curves')
    axes[1].legend()
    axes[1].grid(True)

    plt.tight_layout()
    plt.savefig('model_comparison_curves.png', dpi=300, bbox_inches='tight')
    plt.show()


def main():
    """Main execution function"""
    print("="*80)
    print("LEGAL CLAUSE SIMILARITY DETECTION")
    print("Deep Learning Assignment 2 - NUCES FAST")
    print("="*80)

    # Step 1: Load data
    data_loader = DataLoader()
    clauses, labels = data_loader.load_data()

    # Step 2: Create pairs
    positive_pairs, negative_pairs = data_loader.create_pairs()

    # Combine pairs
    all_pairs = positive_pairs + negative_pairs
    np.random.shuffle(all_pairs)

    print(f"\nTotal pairs: {len(all_pairs)}")
    print(f"Positive pairs: {len(positive_pairs)}")
    print(f"Negative pairs: {len(negative_pairs)}")

    # Step 3: Split data
    train_pairs, temp_pairs = train_test_split(
        all_pairs, test_size=Config.TEST_SPLIT + Config.VALIDATION_SPLIT, random_state=42
    )
    val_pairs, test_pairs = train_test_split(
        temp_pairs, test_size=Config.TEST_SPLIT / (Config.TEST_SPLIT + Config.VALIDATION_SPLIT),
        random_state=42
    )

    print(f"\nTrain pairs: {len(train_pairs)}")
    print(f"Validation pairs: {len(val_pairs)}")
    print(f"Test pairs: {len(test_pairs)}")

    # Step 4: Preprocess text
    preprocessor = TextPreprocessor()

    # Prepare texts for tokenization
    all_texts = [pair[0] for pair in all_pairs] + [pair[1] for pair in all_pairs]
    cleaned_texts = [data_loader.clean_text(text) for text in all_texts]

    preprocessor.fit_tokenizer(cleaned_texts)

    # Prepare sequences
    train_texts1 = [data_loader.clean_text(pair[0]) for pair in train_pairs]
    train_texts2 = [data_loader.clean_text(pair[1]) for pair in train_pairs]
    train_labels = np.array([pair[2] for pair in train_pairs])

    val_texts1 = [data_loader.clean_text(pair[0]) for pair in val_pairs]
    val_texts2 = [data_loader.clean_text(pair[1]) for pair in val_pairs]
    val_labels = np.array([pair[2] for pair in val_pairs])

    test_texts1 = [data_loader.clean_text(pair[0]) for pair in test_pairs]
    test_texts2 = [data_loader.clean_text(pair[1]) for pair in test_pairs]
    test_labels = np.array([pair[2] for pair in test_pairs])

    train_seq1 = preprocessor.texts_to_sequences(train_texts1)
    train_seq2 = preprocessor.texts_to_sequences(train_texts2)
    val_seq1 = preprocessor.texts_to_sequences(val_texts1)
    val_seq2 = preprocessor.texts_to_sequences(val_texts2)
    test_seq1 = preprocessor.texts_to_sequences(test_texts1)
    test_seq2 = preprocessor.texts_to_sequences(test_texts2)

    # Step 5: Train BiLSTM Model
    print("\n" + "="*80)
    print("MODEL 1: BiLSTM Siamese Network")
    print("="*80)

    # Check device placement
    with tf.device('/GPU:0' if tf.config.list_physical_devices('GPU') else '/CPU:0'):
        device_name = 'GPU' if tf.config.list_physical_devices('GPU') else 'CPU'
        print(f"Building model on {device_name}...")

        bilstm_model = BiLSTMSiameseNetwork(
            vocab_size=preprocessor.vocab_size,
            embedding_dim=Config.EMBEDDING_DIM,
            lstm_units=Config.LSTM_UNITS,
            max_seq_length=Config.MAX_SEQUENCE_LENGTH
        )
        bilstm_model.build_model()
        bilstm_model.compile_model()

    print("\nModel Architecture:")
    bilstm_model.model.summary()

    bilstm_trainer = ModelTrainer(bilstm_model.model, "BiLSTM Siamese Network")
    bilstm_training_info = bilstm_trainer.train(
        X_train=(train_seq1, train_seq2),
        y_train=train_labels,
        X_val=(val_seq1, val_seq2),
        y_val=val_labels
    )

    bilstm_results = bilstm_trainer.evaluate(
        X_test=(test_seq1, test_seq2),
        y_test=test_labels
    )
    bilstm_results['training_time'] = bilstm_training_info['training_time']

    bilstm_trainer.plot_training_history('bilstm_training_history.png')

    # Step 6: Train Attention-based Model
    print("\n" + "="*80)
    print("MODEL 2: Attention-based Encoder Network")
    print("="*80)

    # Check device placement
    with tf.device('/GPU:0' if tf.config.list_physical_devices('GPU') else '/CPU:0'):
        device_name = 'GPU' if tf.config.list_physical_devices('GPU') else 'CPU'
        print(f"Building model on {device_name}...")

        attention_model = AttentionEncoderNetwork(
            vocab_size=preprocessor.vocab_size,
            embedding_dim=Config.EMBEDDING_DIM,
            lstm_units=Config.LSTM_UNITS,
            max_seq_length=Config.MAX_SEQUENCE_LENGTH
        )
        attention_model.build_model()
        attention_model.compile_model()

    print("\nModel Architecture:")
    attention_model.model.summary()

    attention_trainer = ModelTrainer(attention_model.model, "Attention-based Encoder")
    attention_training_info = attention_trainer.train(
        X_train=(train_seq1, train_seq2),
        y_train=train_labels,
        X_val=(val_seq1, val_seq2),
        y_val=val_labels
    )

    attention_results = attention_trainer.evaluate(
        X_test=(test_seq1, test_seq2),
        y_test=test_labels
    )
    attention_results['training_time'] = attention_training_info['training_time']

    attention_trainer.plot_training_history('attention_training_history.png')

    # Step 7: Compare models
    compare_models(bilstm_results, attention_results, "BiLSTM", "Attention-based")
    plot_roc_curves(bilstm_results, attention_results, "BiLSTM", "Attention-based")

    # Step 8: Qualitative examples
    bilstm_examples = create_qualitative_examples(
        bilstm_model.model, preprocessor, test_pairs, num_examples=5
    )
    print_qualitative_examples(bilstm_examples, "BiLSTM Siamese Network")

    attention_examples = create_qualitative_examples(
        attention_model.model, preprocessor, test_pairs, num_examples=5
    )
    print_qualitative_examples(attention_examples, "Attention-based Encoder")

    print("\n" + "="*80)
    print("TRAINING COMPLETE!")
    print("="*80)
    print("\nGenerated files:")
    print("  - bilstm_training_history.png")
    print("  - attention_training_history.png")
    print("  - model_comparison_curves.png")


if __name__ == "__main__":
    main()

```
